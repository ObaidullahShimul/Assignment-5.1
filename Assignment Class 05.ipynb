```python
import pandas as pd
import numpy as np
df=pd.read_csv("Feature Transformation.csv")
```

```python
df.head()
```

# 1) Normalization

```python
min_Price=df.Price.min()
```

```python
max_Price=df.Price.max()
```

```python
min_Price
```

```python
max_Price
```

```python
df['Norma_Manual']=(df.Price-min_Price)/(max_Price-min_Price)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms=MinMaxScaler()
```

```python
df['Norma_skLearn']=mms.fit_transform(df[['Price']])
```

```python
df
```

# 2) Standardization

```python
mean_price=np.mean(df['Price'])
```

```python
mean_price
```

```python
std_price=df.Price.std(ddof=0)
```

```python
std_price
```

```python
df['Stand_Manual']=(df.Price-mean_price)/std_price
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df['Stand_skLearn']=sc.fit_transform(df[['Price']])
```

```python
df
```

# 3) Log Transformation

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
fc=FunctionTransformer(np.log1p)
```

```python
df['FC_Manual']=np.log1p(df[['Price']])
df
```

```python
df['FC_skLearn']=fc.fit_transform(df[['Price']])
df
```

# 4) Robust Scaler

```python
Q75,Q25=np.percentile(df.Price,[75,25])
```

```python
Q75
```

```python
Q25
```

```python
IQR=Q75-Q25
IQR
```

```python
price_Med=np.median(df[['Price']])
price_Med
```

```python
df['RS_Manual']=(df[['Price']]-price_Med)/IQR
df
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
robust_sca=RobustScaler(quantile_range=(25.0, 75.0), with_centering=True)
```

```python
df['RS_skLearn']=robust_sca.fit_transform(df[['Price']])
```

```python
df
```

# 5) Max Absolute Scaler

```python
df['Max_Abso_Manual'] = df.Price / df.Price.abs().max()
```

```python
df
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
max_ab_s=MaxAbsScaler()
```

```python
df['Max_Abso_skLearn']=max_ab_s.fit_transform(df[['Price']])
```

```python
df
```
